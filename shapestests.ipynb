```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from circle import Circle
from rectangle import Rectangle
```

```python
    # Visualization method using matplotlib
def plot(self):
    fig, ax = plt.subplots()
    circle = plt.Circle((self._x, self._y), self._radius, color='blue', fill=False, linewidth=2)
    ax.add_artist(circle)

    # Set the aspect ratio to be equal, so the circle isn't distorted
    ax.set_aspect('equal', 'box')

    # Set the limits of the plot
    ax.set_xlim(self._x - self._radius - 1, self._x + self._radius + 1)
    ax.set_ylim(self._y - self._radius - 1, self._y + self._radius + 1)

    # Add labels and grid
    ax.set_title(f"Circle: radius={self._radius}, center=({self._x}, {self._y})")
    ax.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    
    # plt.show()
```

```python
def draw(self, ax):
    """
    Ritar cirkeln i en given matplotlib-ax.
    """
    circle_patch = mpatches.Circle(
        (self._x, self._y),
        self._radius,
        edgecolor='blue',
        facecolor='none',
        linewidth=2
    )
    ax.add_patch(circle_patch)
```

```python
circle = Circle(0, 0, 5)
rectangle = Rectangle(-3, -2, 6, 4)
```

```python
circle.translate(2, 3)
rectangle.translate(-1, -1)
```

```python
#fig, ax = plt.subplots(1, 2, figsize=(16, 4), dpi=200)  

fig, ax = plt.subplots()
circle.draw(ax)
rectangle.draw(ax)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_aspect('equal', adjustable='box')
ax.set_title("Circle with radius 5 at (0,0)")

```
