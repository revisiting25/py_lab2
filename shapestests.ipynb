```python
def test_circle_area():
    circle = Circle(0, 0, 5)   # radius = 5
    expected_area = math.pi * 25  # π * r²
    assert math.isclose(circle.area, expected_area, rel_tol=1e-9)

def test_circle_perimeter():
    circle = Circle(0, 0, 5)   # radius = 5
    expected_perimeter = 2 * math.pi * 5  # 2 * π * r
    
    assert math.isclose(circle.perimeter, expected_perimeter, rel_tol=1e-3) 

def test_is_unit_circle():
    circle = Circle(0, 0, 1)   # radius = 1
    assert circle.is_unit_circle() == True

def test_is_not_unit_circle():
    circle = Circle(0, 0, 2)   # radius = 2
    assert circle.is_unit_circle() == False 

def test_equal_circles():
    circle1 = Circle(0, 0, 5)
    circle2 = Circle(0, 0, 5)
    assert circle1 == circle2
def test_unequal_circles():
    circle1 = Circle(0, 0, 5)
    circle2 = Circle(1, 1, 6)
    assert circle1 != circle2   
def test_lt_circle_radius():
    circle1 = Circle(0, 0, 3)
    circle2 = Circle(0, 0, 5)
    assert circle1 < circle2    


test_circle_area()
test_circle_perimeter()
test_is_unit_circle()
test_is_not_unit_circle()
test_equal_circles()
test_unequal_circles()
test_lt_circle_radius()
```

```python
rectangle_test = Rectangle(-1, -1, 4, 2)
print("Representation: " + rectangle_test.__repr__())
print('Rectangle(bottom_left=(-1, -1), width=4, height=2)')
print("String: " + rectangle_test.__str__())
print('Rectangle with width 4 and height 2 at (-1, -1)')        

print("Width: " + str(rectangle._width))
print("Height: " + str(rectangle._height))
print("Area: " + str(rectangle.area))
print("Perimeter: " + str(rectangle.perimeter))
```

```python
def test_rectangle_area():
    rectangle = Rectangle(0, 0, 4, 5)   # width = 4, height = 5
    expected_area = 4 * 5  # width * height
    assert math.isclose(rectangle.area, expected_area, rel_tol=1e-9)    
def test_rectangle_perimeter():
    rectangle = Rectangle(0, 0, 4, 5)   # width = 4, height = 5
    expected_perimeter = 2 * (4 + 5)  # 2 * (width + height)
    assert math.isclose(rectangle.perimeter, expected_perimeter, rel_tol=1e-3) 
def test_is_square():
    rectangle = Rectangle(0, 0, 4, 4)   # width = 4, height = 4
    assert rectangle.is_square() == True
def test_is_not_square():
    rectangle = Rectangle(0, 0, 4, 5)   # width = 4, height = 5
    assert rectangle.is_square() == False 
def test_equal_rectangles():
    rectangle1 = Rectangle(0, 0, 4, 5)
    rectangle2 = Rectangle(0, 0, 4, 5)
    assert rectangle1 == rectangle2
def test_unequal_rectangles():
    rectangle1 = Rectangle(0, 0, 4, 5)
    rectangle2 = Rectangle(1, 1, 5, 6)
    assert rectangle1 != rectangle2   
def test_lt_rectangle_area():
    rectangle1 = Rectangle(0, 0, 3, 4)  # area = 12
    rectangle2 = Rectangle(0, 0, 4, 5)  # area = 20
    assert rectangle1 < rectangle2  
```

```python
test_rectangle_area()
test_rectangle_perimeter()
test_is_square()
test_is_not_square()
test_equal_rectangles()
test_unequal_rectangles()
test_lt_rectangle_area()    
```
